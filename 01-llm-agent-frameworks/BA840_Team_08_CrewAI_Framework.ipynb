In this notebook, we used CrewAI to create an agent that plans travel itineraries. Our 'crew' consists of specialized agents that handle different aspects of the travel planning process. The first agent gathers travel preferences, the second suggests destinations and activities, the third generates detailed travel itineraries.

```python
!pip install crewai==0.28.8 crewai_tools==0.1.6 langchain_community==0.0.29
!pip install utils

from crewai import Agent, Task, Crew
import os
```

```python
!pip install python-dotenv

from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

# Access API key from environment variable
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    raise ValueError("OPENAI_API_KEY not found in environment variables. Please create a .env file with your key.")
```

```python
from crewai import Agent, Task, Crew
import os

# Fetch your OpenAI API key
os.environ["OPENAI_API_KEY"] = api_key
os.environ["OPENAI_MODEL_NAME"] = 'gpt-3.5-turbo'

# Define agents for the travel itinerary chatbot
preferences_agent = Agent(
    role='Gather Travel Preferences',
    goal='Ask the user about their travel preferences for a trip to Europe',
    backstory='You are a friendly travel agent.',
    allow_delegation=False,
    verbose=True
)

suggestion_agent = Agent(
    role='Suggest Destinations and Activities',
    goal='Suggest suitable travel destinations and activities based on the preferences of a group of 4 grad students traveling on a budget',
    backstory='You are an experienced travel guide.',
    allow_delegation=False,
    verbose=True
)

itinerary_agent = Agent(
    role='Generate Detailed Travel Itinerary',
    goal='Create a detailed travel itinerary for a week-long summer trip to Europe, including travel days, for a group of 4 grad students traveling on a budget',
    backstory='You are an expert in planning travel itineraries.',
    allow_delegation=False,
    verbose=True
)

# Define tasks for each agent
gather_preferences = Task(
    description='Ask the user about their travel preferences, such as preferred destinations, budget, and activities for their summer trip to Europe.',
    expected_output='User travel preferences',
    agent=preferences_agent
)

suggest_destinations = Task(
    description='Based on the gathered preferences, suggest suitable travel destinations and activities for a group of 4 grad students traveling on a budget.',
    expected_output='List of suggested destinations and activities',
    agent=suggestion_agent
)

generate_itinerary = Task(
    description='Create a detailed travel itinerary for a week-long summer trip to Europe, including travel days, based on the selected destination and user preferences.',
    expected_output='Travel itinerary',
    agent=itinerary_agent
)

# Set up the crew with the agents and tasks
crew = Crew(
    agents=[preferences_agent, suggestion_agent, itinerary_agent],
    tasks=[gather_preferences, suggest_destinations, generate_itinerary],
    verbose=2
)
```

```python
travel_prompt = """
Plan a detailed travel itinerary for a summer trip to Europe for a group of 4 grad students traveling on a budget.
It will be a week-long trip including travel days.
The students are currently in Boston, United States, and are looking to travel in June.
"""

result = crew.kickoff(inputs={'travel_prompt': travel_prompt})

print("Your travel itinerary is ready:")
print(result)
```
